```python
import matplotlib.pyplot as plt

# Sample data
products = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
sales = [50, 75, 30, 90, 60]

# Create a bar chart
plt.bar(products, sales, color='black')

# Add labels and title
plt.xlabel('Products')
plt.ylabel('Sales (Units)')
plt.title('Product Sales Visualization')

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data (e.g., ages between 18 and 60)
ages = np.random.randint(18, 60, 100)

# Create histogram
plt.hist(ages, bins=8, color='lightgreen', edgecolor='black')

# Add labels and title
plt.xlabel('Age Range')
plt.ylabel('Number of People')
plt.title('Age Distribution Histogram')

# Show the chart
plt.show()
```
